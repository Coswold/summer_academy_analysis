```python
def map_types_to_value(df, column_name, type_to_change, change_to_value):
    '''Takes a Pandas Dataframe, column name, object type to change, and the value to replace with. Returns a "mapped" dataframe based on value types that you want to replace'''
    for index, row in df.iterrows():
        changer = None
        if row[column_name] == type_to_change:
            changer = change_to_value
            df.set_value(index, column_name, changer)
    return df
        
frame = map_types_to_value(frame, "Schedule Pacing", "Just right", 5)
frame = map_types_to_value(frame, "Schedule Pacing", "Way too fast", 1)
frame = map_types_to_value(frame, "Schedule Pacing", "Way too slow", 1)
frame = map_types_to_value(frame, "Schedule Pacing", "A little too fast", 3)
frame = map_types_to_value(frame, "Schedule Pacing", "A little too slow", 3)
frame = map_types_to_value(frame, "Schedule Pacing", "#REF!", 0)
frame = map_types_to_value(frame, "How well is the schedule paced?", "#REF!", 0)
frame = map_types_to_value(frame, "How well are the tutorials paced?", "#REF!", 0)


```
