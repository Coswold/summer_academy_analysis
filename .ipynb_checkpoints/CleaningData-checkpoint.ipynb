```python
import numpy as np
import pandas as pd
import glob


df = pd.read_csv("SA/2016/Anon Week 1 Feedback - Singapore.csv")
```

```python
path =r'SA/2016' 
path2 =r'SA/2017'
allFiles = glob.glob(path + "/*.csv")
allFiles += glob.glob(path2 + "/*.csv")

list_ = []

for file_ in allFiles:
    df = pd.read_csv(file_,index_col=None, header=0)
    list_.append(df)

frame = pd.concat(list_, axis = 0, ignore_index = True, sort=False)
```

```python
frame
```

```python
frame = frame.drop(columns = ['Timestamp', 'Start Date (UTC)', 'Submit Date (UTC)', 'Network ID', 'Unnamed: 0', 'ID', '#'])
```

```python
frame
```

```python
frame[['location', 'Location']] = frame[['location','Location']].fillna(value=0)
```

```python
frame["location"] = frame["location"].map(str) + frame["Location"].map(str)
```

```python
frame = frame.drop(columns=['Location'])
```

```python
frame
```

```python
frame[['track', 'Track', 'What track are you in?']] = frame[['track','Track', 'What track are you in?']].fillna(value=0)
```

```python
frame["track"] = frame["track"].map(str) + frame["Track"].map(str) + frame["What track are you in?"].map(str)
```

```python
frame = frame.drop(columns=['What track are you in?', 'Track'])
```
